Perform experimental analysis to test the hypothesis that Pythonâ€™s sorted method runs in O(nlogn) time on average

```python
import time
import random
import matplotlib.pyplot as plt
import numpy as np

def measure_time(n):
    lst = random.sample(range(n), n)
    start = time.time()
    sorted(lst)
    return time.time() - start

sizes = [2**i for i in range(11, 24)]
times = [measure_time(n) for n in sizes]  #

##Plot actual sorting times
plt.plot(sizes, times, label='Actual Time', marker='o')

nlogn = [n * np.log2(n) / 1e6 for n in sizes]
plt.plot(sizes, nlogn, label='O(n log n)', linestyle='--')

plt.xscale('log')
plt.yscale('log')
plt.xlabel('Input Size (n)')
plt.ylabel('Time (seconds)')
plt.title('Python sorted() Time Complexity')
plt.legend()
plt.grid(True)

plt.show()
```
