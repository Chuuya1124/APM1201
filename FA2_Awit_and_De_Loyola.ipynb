P-2.33 (10 points)

Write a Python program that inputs a polynomial in standard algebraic notation and outputs the first derivative of that polynomial.


```python
import re

class Polynomial:
    def __init__(self, expression):
        self.expression = expression
        self.terms = self.parse_expression()

    def parse_expression(self):

        pattern = r'([+-]?\s*\d*\s*x?\^?\d*)'
        terms = re.findall(pattern, self.expression.replace(" ", ""))
        return [term for term in terms if term]

    def derivative(self):
        derived_terms = []
        for term in self.terms:

            if 'x' in term:
                if '^' in term:
                    coeff, exp = term.split('x^')
                    exp = int(exp)
                    coeff = int(coeff) if coeff and coeff != '+' and coeff != '-' else (1 if coeff == '+' or coeff == '' else -1)
                    new_coeff = coeff * exp
                    new_exp = exp - 1
                    if new_exp > 0:
                        derived_terms.append(f"{new_coeff}x^{new_exp}")
                    else:
                        derived_terms.append(f"{new_coeff}")
                else:
                    coeff = term[:-1]
                    coeff = int(coeff) if coeff and coeff != '+' and coeff != '-' else (1 if coeff == '+' or coeff == '' else -1)
                    derived_terms.append(f"{coeff}")
            else:

                continue

        return " + ".join(derived_terms).replace(" + -", " - ").replace(" - ", " - ")

def main():
    poly_input = input("Enter a polynomial: ")
    polynomial = Polynomial(poly_input)
    print("First derivative:", polynomial.derivative())

if __name__ == "__main__":
    main()
```

Develop an inheritance hierarchy based upon a Polygon class that has abstract methods area( ) and perimeter( ). Implement classes Triangle, Quadrilateral, Pentagon, Hexagon, and Octagon that extend this base class, with the obvious meanings for the area( ) and perimeter( ) methods. Also implement classes, IsoscelesTriangle, EquilateralTriangle, Rectangle, and Square, that have the appropriate inheritance relationships. Finally, write a simple program that allows users to create polygons of the various types and input their geometric dimensions, and the program then outputs their area and perimeter. For extra effort, allow users to input polygons by specifying their vertex coordinates and be able to test if two such polygons are similar

```python
from abc import ABC, abstractmethod
import math

class Polygon(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

class Triangle(Polygon):
    def __init__(self, a, b, c):
        self.sides = [a, b, c]

    def area(self):
        s = self.perimeter() / 2
        return math.sqrt(s * (s - self.sides[0]) * (s - self.sides[1]) * (s - self.sides[2]))

    def perimeter(self):
        return sum(self.sides)

class IsoscelesTriangle(Triangle):
    def __init__(self, a, b):
        super().__init__(a, b, b)

class EquilateralTriangle(IsoscelesTriangle):
    def __init__(self, a):
        super().__init__(a, a)

class Quadrilateral(Polygon):
    def __init__(self, a, b, c, d):
        self.sides = [a, b, c, d]

    def area(self):
        s = self.perimeter() / 2
        return math.sqrt((s - self.sides[0]) * (s - self.sides[1]) * (s - self.sides[2]) * (s - self.sides[3]))

    def perimeter(self):
        return sum(self.sides)

class Rectangle(Quadrilateral):
    def __init__(self, width, height):
        super().__init__(width, height, width, height)

    def area(self):
        return self.sides[0] * self.sides[1]

class Square(Rectangle):
    def __init__(self, side):
        super().__init__(side, side)

class Pentagon(Polygon):
    def __init__(self, sides):
        self.sides = sides

    def area(self):

        s = self.perimeter() / 5
        return (1/4) * math.sqrt(5 * (5 + 2 * math.sqrt(5))) * (s ** 2)

    def perimeter(self):
        return sum(self.sides)

class Hexagon(Polygon):
    def __init__(self, sides):
        self.sides = sides

    def area(self):

        s = self.perimeter() / 6
        return (3 * math.sqrt(3) / 2) * (s ** 2)

    def perimeter(self):
        return sum(self.sides)

class Octagon(Polygon):
    def __init__(self, sides):
        self.sides = sides

    def area(self):

        s = self.perimeter() / 8
        return 2 * (1 + math.sqrt(2)) * (s ** 2)

    def perimeter(self):
        return sum(self.sides)

def are_similar(polygon1, polygon2):
    if isinstance(polygon1, Polygon) and isinstance(polygon2, Polygon):
        return sorted(polygon1.sides) == sorted(polygon2.sides)
    return False

def main():
    while True:
        print("Choose a polygon to create:")
        print("1. Triangle")
        print("2. Isosceles Triangle")
        print("3. Equilateral Triangle")
        print("4. Rectangle")
        print("5. Square")
        print("6. Pentagon")
        print("7. Hexagon")
        print("8. Octagon")
        print("9. Exit")

        choice = input("Enter your choice: ")

        if choice == '9':
            break

        if choice in ['1', '2', '3']:
            sides = list(map(float, input("Enter the lengths of the sides separated by spaces: ").split()))
            if choice == '1':
                polygon = Triangle(*sides)
            elif choice == '2':
                polygon = IsoscelesTriangle(*sides)
            else:
                polygon = EquilateralTriangle(*sides)

        elif choice in ['4', '5']:
            sides = list(map(float, input("Enter the width and height separated by spaces: ").split()))
            if choice == '4':
                polygon = Rectangle(*sides)
            else:
                polygon = Square(sides[0])

        elif choice in ['6', '7', '8']:
            sides = list(map(float, input("Enter the lengths of the sides separated by spaces: ").split()))
            if choice == '6':
                polygon = Pentagon(sides)
            elif choice == '7':
                polygon = Hexagon(sides)
            else:
                polygon = Octagon(sides)

        else:
            print("Invalid choice.")
            continue

        print(f"Area: {polygon.area()}")
        print(f"Perimeter: {polygon.perimeter()}")

        check_similar = input("Do you want to check similarity with another polygon? (yes/no): ")
        if check_similar.lower() == 'yes':
            other_sides = list(map(float, input("Enter the lengths of the sides of the other polygon separated by spaces: ").split()))
            if len(other_sides) == 3:
                other_polygon = Triangle(*other_sides)
            elif len(other_sides) == 4:
                other_polygon = Quadrilateral(*other_sides)
            elif len(other_sides) == 5:
                other_polygon = Pentagon(other_sides)
            elif len(other_sides) == 6:
                other_polygon = Hexagon(other_sides)
            elif len(other_sides) == 8:
                other_polygon = Octagon(other_sides)
            else:
                print("Invalid number of sides for similarity check.")
                continue

            print("Are the polygons similar?", are_similar(polygon, other_polygon))

if __name__ == "__main__":
    main()

```
